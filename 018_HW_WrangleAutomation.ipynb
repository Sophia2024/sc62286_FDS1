```python
def automation_wrangling(filename, var1 = ['Condition_A', 'Condition_B'], var2 = ['level_1', 'level_2']) :
    import pandas as pd
    import numpy as np
    myFile = pd.read_csv(filename)
    
    # convert to numpy array
    myData = myFile.to_numpy()
    
    # get the numbers of rows and columns 
    obs, grp = myData.shape
    
    new_length = obs * grp
    
    values_col = np.reshape(myData, (new_length, 1), order = "F")
    
    values_col = np.squeeze(values_col)
    
    # constructing the outer grouping variable
    condition = pd.Series(var1)
    condition = condition.repeat(obs * 2)
    condition = condition.reset_index(drop = True)
    
    # constructing the innner grouping variable
    level = pd.Series(var2)
    level = level.repeat(obs)
    level = pd.concat([level] * 2, ignore_index = True)
    
    # constructing the data frame
    dataFrame = pd.DataFrame(
        {
        "Data" : values_col, 
        "Condition" : condition,
        "Level" : level
        }
    )
    
    return dataFrame
    
    
```

convert from wide to tidy long format
