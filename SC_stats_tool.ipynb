```python

instr = "restart"
while instr != "quit" :
    import SC_stats_module as m
    m.menu()
    
    # the first round, enter number, done, enter instruction, compute w/ if elif else (exclude invalid input)
     
    if instr == "restart" :
        all_nums = []
        num = input("Please enter a number: ")
        while num != "done" : # if num is not "done", keep asking user for next number
            # judge if num is a valid number or some other things 
            if not m.is_numeric(num) : # num is not a valid number
                num = input("Error Message: Please enter a VALID number: ")
            else : 
                # num is a valid number
                num = float(num) # turn the num into a numeric
                all_nums.append(num) # create a list of all the numbers entered by the user [floats]
                num = input("Please enter a number: ") # ask for the next number
            # num == "done", while loop stops
        print("Message: You are done entering the numbers! \n")
        
        # enter instruction for computation
        instr = input("Please enter the instruction number: ")
        while instr != "restart" :
            if instr == "1" :
                mean = m.mean(all_nums)
                print("Result: The mean of all the numbers is:", mean)
                instr = input("Please enter the next instruction: ")
            elif instr == "2" :
                variance = m.variance(all_nums)
                print("Result: The variance of all the numbers is:", variance)
                instr = input("Please enter the next instruction: ")
            elif instr == "3" :
                sd = m.sd(all_nums)
                print("Result: The standard deviation of all the numbers is:", sd)
                instr = input("Please enter the next instruction: ")
            elif instr == "4" :
                se = m.se(all_nums)
                print("Result: The standard error of all the numbers is:", se)
                instr = input("Please enter the next instruction: ")
            elif instr == "5" :
                observation = input("Please enter the number in order to compute the correspondent z-score: ")
                # if observation is not a valid number
                while not m.is_numeric(observation) :
                    observation = input("Error Message: Please enter a valid number: ")
                observation = float(observation)
                zscore = m.zscore(observation, all_nums)
                print("Result: The z-score for the number", observation,"is:", zscore)
                instr = input("Please enter the next instruction: ")
            elif instr == "6" :
                observation = input("Please enter the number in order to compute the correspondent z-score: ")
                # if observation is not a valid number
                while not m.is_numeric(observation) :
                    observation = input("Error Message: Please enter a valid number: ")
                observation = float(observation)
                m.summary(observation, all_nums)
                instr = input("Please enter the next instruction: ")
            elif instr == "quit" :
                break 
                
            else : # instr is a invalid one
                instr = input("Error Message: Please enter a valid instruction number from 1 to 5: ")
        # instr == restart, restart from enter number, and then perform everything all over again
        
# instr == quit, quit the program
print("You successfully quit the program!")

```
